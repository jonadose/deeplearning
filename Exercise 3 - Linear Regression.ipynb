## Downloading and loading the dataset

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
import seaborn as sns 

df = pd.read_csv('auto.csv')
#removing the '?'
df = df[df.horsepower != '?']
df
# convert dtype to int
df = df.astype({"horsepower": 'int64'})
df.dtypes

#Mantas sns magic
sns.heatmap(df.corr(), 
            xticklabels=df.corr().columns.values,
            yticklabels=df.corr().columns.values)
```

## Inspecting the dataset 

```python
sns.set(style="whitegrid")
sns_plot = sns.pairplot(df, hue = "origin")
sns_plot.savefig("mpg_graph.png")
```

### useful parameters to map: 

- weight to mpg 
- horsepower to mpg 
- displacement to mpg 
- model_year to mpg
- cylinders to mpg

## Performing linear regression 

```python
x_horsepower = df['horsepower']
y_mpg = df['mpg']

## somehow you have to add a constant to the x 
x_horsepower = sm.add_constant(x_horsepower)

model = sm.OLS(y_mpg, x_horsepower)
results = model.fit()
results.params
print(results.summary())
print('Parameters: ', results.params)
print('R2: ', results.rsquared)
```

## Plotting the results 

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(results, fig=fig)

fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.plot_fit(results, 1, ax=ax)
plt.show()
```

## Extending to all Features 

```python
x = df[["horsepower","displacement", "model_year", "cylinders", "origin"]]
y_mpg = df['mpg']

## somehow you have to add a constant to the x 
x = sm.add_constant(x)

model_all = sm.OLS(y_mpg, x)
results_all = model_all.fit()
results_all.params
print(results_all.summary())
print('Parameters: ', results_all.params)
print('R2: ', results_all.rsquared)
```

## Plotting it all 

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(results_all, fig=fig)

fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.plot_fit(results_all, 1, ax=ax)
plt.show()
```

## Refining the model based on correlation values - cut everything with p-value above 0.05

```python
# removed horsepower and cylinders
x = df["weight"]
y_mpg = df['mpg']

## somehow you have to add a constant to the x 
x = sm.add_constant(x)

model_ref = sm.OLS(y_mpg, x)
results_ref = model_ref.fit()
results_ref.params
print(results_ref.summary())
print('Parameters: ', results_ref.params)
print('R2: ', results_ref.rsquared)
```

## Plotting refined graph

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(results_ref, fig=fig)

fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.plot_fit(results_ref, 1, ax=ax)
plt.show()
```

## Conclusion the worst regressor is (model year) - weight (np.sqrt(weight))
